# Development Anonymization

```python
import pandas as pd
train_path = r'/Users/tom/Desktop/train_indiv.csv'
test_path = r'/Users/tom/Desktop/test_indiv.csv'
train = pd.read_csv(train_path, index_col=0, encoding='utf-8')
test = pd.read_csv(test_path, index_col=0, encoding='utf-8')
```

```python
usernames = list(set(list(set(train['Username'].values)) + list(set(test['Username'].values))))
```

```python
country = 'UK'
df = train
df_name = 'train'
user2anon = {user:'Dev_{}_User_{}'.format(country, ind) for ind, user in enumerate(usernames)}
new = []


for user in df['Username'].values:
    new.append(user2anon[user])
new = pd.Series(new, index=df['Username'].index)
```

```python
df['Username'] = new
df
```

```python
df.to_csv('{}_Dev_{}_Anon.csv'.format(country, df_name))
```

# Experiment Anonymization

```python
import pandas as pd
path = r'/Users/tom/Desktop/indiv_experiment.csv'
data = pd.read_csv(path, index_col=0, encoding='utf-8')
usernames = list(set(data['Username'].values))
```

```python
country = 'Spain'

user2anon = {user:'Exp_{}_User_{}'.format(country, ind) for ind, user in enumerate(usernames)}
new = []


for user in data['Username'].values:
    new.append(user2anon[user])
new = pd.Series(new, index=data['Username'].index)
data['Username'] = new
data
```

```python
data.to_csv('Exp_{}_Anon.csv'.format(country))
```
