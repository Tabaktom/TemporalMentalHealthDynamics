```python
import re
import pandas as pd
import os
from tqdm import tqdm
origin = r'/Users/tom/Desktop/new_depression_clean/'
destination = r'/Users/tom/Desktop/new_depression_clean_clean/'

origin = r'/Users/tom/Desktop/Italy_tweets/development/diagnosed_depression_initial_clean/'
destination = r'/Users/tom/Desktop/Italy_tweets/development/diagnosed_depression_clean/'

origin=r'/Users/tom/Desktop/spain_tweets/development/diagnosed_depression_tweets/'
destination=r'/Users/tom/Desktop/spain_tweets/development/diagnosed_depression_clean/'
origin=r'/Users/tom/Desktop/spain_tweets/experiment/pre_lockdown/'
destination=r'/Users/tom/Desktop/spain_tweets/experiment/pre_lockdown_clean/'

origin=r'/Users/tom/Desktop/France_tweets/development/control_group/'
destination=r'/Users/tom/Desktop/France_tweets/development/control_group_clean/'

origin = r'/Users/tom/Desktop/Italy_tweets/experiment/during/'
destination = r'/Users/tom/Desktop/Italy_tweets/experiment/during_clean/'

origin=r'/Users/tom/Desktop/France_tweets/experiment/pre_lockdown/'
destination =r'/Users/tom/Desktop/France_tweets/experiment/pre_clean/'

origin=r'/Users/tom/Desktop/germany_tweets/development/diagnosed/'
destination=r'/Users/tom/Desktop/germany_tweets/development/diagnosed_clean/'

origin = r'/Users/tom/Desktop/germany_tweets/development/control/'
destination = r'/Users/tom/Desktop/germany_tweets/development/control_clean/'

origin  =r'/Users/tom/Desktop/germany_tweets/experiment/all/'
destination =r'/Users/tom/Desktop/germany_tweets/experiment/all_clean/'

print(len(os.listdir(origin)))
origin_files=[fname for fname in os.listdir(origin)]
for f in tqdm(origin_files):
    if f != '.DS_Store':
        data = pd.read_csv(origin +f, index_col=0, encoding='utf-8')
        new = []
        punct = ['?', '!', '&', '.', ',', '<', 'S', '/', '>']
        for ind, row in data.iterrows():
            interim = ''.join([i for i in row.CTweet if not i.isdigit() and i not in punct])
            new.append('<S>' + re.sub(r'(?<=\w)[^\s\w](?![^\s\w])', '', interim)+'<S/> ')
        data["CTweet"] = pd.Series(new, index=data.index)
        data.to_csv(destination+f,encoding='utf-8')
```
